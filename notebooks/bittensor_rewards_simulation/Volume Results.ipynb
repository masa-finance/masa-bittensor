```python
import pandas as pd
import matplotlib.pyplot as plt
import json

real_data = json.load(open("volumes.json"))

# Calculate the sum of volumes for each miner
miner_volumes = {}
for data in real_data:
    for miner, volume in data["miners"].items():
        if volume > 0:  # Only consider miners with volume greater than 0
            if miner in miner_volumes:
                miner_volumes[miner] += volume
            else:
                miner_volumes[miner] = volume

# Convert to DataFrame for easier plotting
df = pd.DataFrame(list(miner_volumes.items()), columns=['Miner', 'Volume'])

# Sort the DataFrame by total volume
df = df.sort_values(by='Volume', ascending=False)

# Plot the volumes
plt.figure(figsize=(10, 6))
plt.bar(df['Miner'], df['Volume'])
plt.xlabel('Miner')
plt.ylabel('Total Volume')
plt.title('Total Volume by Miner')
plt.xticks(rotation=90)
plt.show()
```
